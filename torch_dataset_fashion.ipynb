```python
import torchvision
import torch

from torchvision.datasets import FashionMNIST
from torch.utils.data import DataLoader
from torchvision import transforms
```

```python
transform_data = transforms.ToTensor()
```

```python
train_data = FashionMNIST(root= r"C:\Users\abhis\Mine\CV_parle", train= True, download= True, transform= transform_data)
```

```python
test_data = FashionMNIST(root=r"C:\Users\abhis\Mine\CV_parle", train= False, download= True, transform= transform_data)
```

```python
print("Train Data: ", len(train_data))
print("Test Data:  ", len(test_data))
```

```python
import math
batch_size = 30
train_loader = DataLoader(train_data,batch_size=30, shuffle=True)
test_loader = DataLoader(test_data, batch_size=30, shuffle=True)
no_of_columns = math.ceil(batch_size/2)
```

```python
classes = ["T-shirts", "Trouser", "PullOver", "Dress","Coat", 
          "Sanadal", "Shirt", "Sneaker", "Bag", "Ankle Boot"]
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
for i in train_loader:
    images, labels = i
images = images.numpy()

fig = plt.figure(figsize=(12,4))
for i in np.arange(batch_size):
    ax = fig.add_subplot(2,no_of_columns,i+1, xticks = [], yticks = [])
    ax.imshow(np.squeeze(images[i]),cmap = "gray")
    ax.set_title(classes[labels[i]])
```
