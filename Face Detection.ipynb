```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline
```

```python
image = cv2.imread(r"C:\Users\abhis\Mine\CV_parle\images\group_image.jpg")
plt.imshow(image)
```

```python
image_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
```

```python
face_cascade = cv2.CascadeClassifier(r"C:\Users\abhis\Mine\CV_parle\haarcascade_frontalface_default.xml")
```

```python
faces = face_cascade.detectMultiScale(gray, 4,6)
```

```python
print ('We found ' + str(len(faces)) + ' faces in this image')
print ("Their coordinates and lengths/widths are as follows")
print ('=============================')
print (faces)
```
